## Descubrir temas 

#### Objetivo: 
Analizar todas las frases y preguntas del usuario que el bot no supo responder con el objetivo de descubrir nuevos tópicos de interés o asignar las frases a tópicos ya existentes.

Para éste análisis se consideraron las conversaciones de los últimos 14 días a la fecha de hoy. Se han tomado los registros de las conversaciones y se han seleccionado aquellas frases del usuario para las cuales el bot respondió: "Perdón, no sé si entendí correctamente" y a continuación un pedido de elegir la pregunta probable mediante botones. Éstas están en el archivo new_intents.txt

### LDA y NMF


```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
import matplotlib.pyplot as plt
import collections
import statistics as stats
from nltk.corpus import stopwords
from wordcloud import WordCloud
```

Se cargan las frases no respondidas y se preprocesan para eliminar tildes:

```python
def preprocessing(doc):
    doc = doc.lower()
    doc = doc.replace('á','a').replace('é','e').replace('í','i').replace('ó','o').replace('ú','u')
    return doc

documents = []
with open('./new_intents.txt','rt', encoding='utf8') as f:
    for line in f:
        documents.append(line.rstrip('\n'))

documents = [preprocessing(doc) for doc in documents]
len(documents)
```

Las stopwords son aquellas palabras con una alta frecuencia de aparición en un idioma, pero que no tienen mucho significado en si mismas. Por ejemplo: artículos, conjunciones, preposiciones, etc. 

En éste caso decido buscar las stopwords específicas para el corpus y no considerar una lista precargada. Para ello se hace un conteo de todas las palabras y se sacan medidas estadísticas sobre las frecuencias obtenidas.

```python
words_corpus = (' '.join(documents)).split()
frecs = collections.Counter(words_corpus)

print('minimo: ',min(frecs.values()))
print('media: ',round(stats.mean(frecs.values()),2))
print('maximo: ',max(frecs.values()))
print('mediana: ',stats.median(frecs.values()))
```

Se ve que hasta la mitad de las palabras aparecen una vez en todo el corpus, mientras que la frecuencia máxima es de 32. Viéndolo visualente:

```python
plt.figure(figsize=(8,4))
plt.grid(color='w', linestyle='solid')
plt.hist(list(frecs.values()), bins=15, density=False)
plt.show()
```

Se inspeccionan aquellas palabras con frecuencia mayor a 6:

```python
stopws = [k for k,v in frecs.items() if v>6]
stopws
```

De éstas eliminaría la palabra córdoba por considerar que si aporta significado:

```python
stopws.pop(-1)
```

Se vectorizan las frases. Para ello se usan las clases de sklearn que arman una matriz donde cada fila se corresponde con un documento y cada columna con un token o feature.

En el caso de tf-idf se pesa la matriz de frecuencias de modo de atenuar el peso de los términos que ocurren con alta frecuencia en toda la colección de documentos en general.

NMF puede usar la matriz tf-idf.

```python
vectorizertfidf = TfidfVectorizer(max_df=0.90, min_df=2, stop_words=stopws)
tfidf = vectorizertfidf.fit_transform(documents)
tfidf_featurenames = vectorizertfidf.get_feature_names()
print("Nro de filas/documentos: %s  \nNro de columnas/tokens %s" % tfidf.toarray().shape)
```

Si se quiere ver un ejemplo de como se estan tokenizando las frases se puede hacer:

```python
analyze = vectorizertfidf.build_analyzer()
analyze("Este es un, doCumento nuevo?. mamá")
```

 LDA solo puede usar el conteo de frecuencias en crudo por ser un modelo probabilístico.

```python
tfvectorizer = CountVectorizer(max_df=0.90, min_df=2, stop_words=stopws)
tf = tfvectorizer.fit_transform(documents)
tf_featurenames = tfvectorizer.get_feature_names()
print("Nro de filas/documentos: %s  \nNro de columnas/tokens %s" % tf.toarray().shape)
```

Un ejemplo de como se tokenizan las frases:

```python
analyze = tfvectorizer.build_analyzer()
analyze("Este es un, doCumento nuevo?. mamá")
```

Se modelan los temas con NMF y LDA y se muestran las palabras más representativas de cada tema:

NMF- Factorización de matrices no negativas. Se busca reducir las dimensiones en los datos y que surjan ciertas características que permanecen ocultas a priori.

LDA- Latent Dirichlet Allocation. Método probabilístico para el modelado de temas.

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
    print()

        
no_topics = 20

# Run NMF
nmf = NMF(n_components=no_topics, random_state=12, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf)

# Run LDA
lda = LatentDirichletAllocation(n_components=no_topics, max_iter=5, learning_method='online', 
                                learning_offset=50., random_state=12).fit(tf)

no_top_words = 5

display_topics(nmf, tfidf_featurenames, no_top_words)
display_topics(lda, tf_featurenames, no_top_words)
```

### Nube de palabras

Es muy conocida la nube de palabras que representa con mayor tamaño aquellas palabras que ocurren con más frecuencia:

```python
stopws = set(stopwords.words('spanish'))

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopws,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1).generate(str(data))
    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)
    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(documents)
```

En ambos casos se pueden ver palabras muy presentes en las frases de los usuarios. Algunos temas podrían ser:
- uso de barbijos
- lugares abiertos
- ferreterías
- cantidad de infectados en córdoba
- tasa de mortalidad
- preguntas locales de córdoba
- los test

En el caso de la nube de palabras se ve que quedan más stopwords por quitar..
